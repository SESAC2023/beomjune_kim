```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv('complete.csv')

# 각 label1 값에 따라 100개씩 랜덤하게 추출하면서 원본 데이터에서 제거
dfs = []
for label in [1, 2, 3, 4, 5]:
    sample_df = df[df['label1'] == label].sample(100, random_state=42)
    dfs.append(sample_df)

    # 원본 데이터에서 추출된 데이터 제거
    df.drop(sample_df.index, inplace=True)

# 추출된 데이터 합치기
result_df = pd.concat(dfs, ignore_index=True)

# 새로운 CSV 파일로 저장
result_df.to_csv('./complete_test.csv', index=False)

# 원본 데이터에서 추출된 데이터를 제거한 후 다시 저장
df.to_csv('./complete_train.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('complete_test.csv')

# label1, label2 열의 값들이 몇 개씩 있는지 확인 후 오름차순 정렬
label1_counts = df['label1'].value_counts().sort_index()
label2_counts = df['label2'].value_counts().sort_index()

# 결과 출력
print("label1 counts:")
print(label1_counts)
print("\nlabel2 counts:")
print(label2_counts)

# 그래프로 나타내기
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

label1_counts.plot(kind='bar', ax=ax[0], title='label1 counts')
label2_counts.plot(kind='bar', ax=ax[1], title='label2 counts')

plt.tight_layout()
plt.show()
```
